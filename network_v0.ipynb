```python
import pickle
import numpy as np
FF = [ 
      ['logdir/dicts/log_uninduced_01.pkl',\
        'logdir/dicts/log_uninduced_02.pkl',\
        'logdir/dicts/log_uninduced_03.pkl',\
        'logdir/dicts/log_uninduced_04.pkl'], \
        ['logdir/dicts/log_induced_01.pkl',\
        'logdir/dicts/log_induced_02.pkl',\
        'logdir/dicts/log_induced_03.pkl',\
        'logdir/dicts/log_induced_04.pkl'],\
        ['logdir/dicts/log_uninduced_CNN.pkl',\
        'logdir/dicts/log_induced_CNN.pkl'] 
    ]        
        
# ------------------------------------------------
        



Data=[]
for files in FF:
    TT = []
    VV= []
    TeTe= []
    for file in files:    
        with open(file, 'rb') as f:
            loaded_dict = pickle.load(f)
        Train = []
        Valid = []
        Test =  []
        print(loaded_dict)        
        for key in loaded_dict.keys():
            dic = loaded_dict[key]
            train = []
            valid = []
            test =  []
            print(len(dic.keys()), key)
            
            for keys in dic.keys():
                # print( np.array(dic[keys]).shape )
                        
                if keys.startswith('train'):
                    train.append( dic[keys])
                if keys.startswith('valid'):
                    valid.append( dic[keys])
                if keys.startswith('test'):
                    test.append( dic[keys])
                    
                    
            Test.append(np.array(test))
            Valid.append(np.array(valid))
            Train.append(np.array(train))
            

        TT.append(np.array(Train))
        VV.append(np.array(Valid))
        TeTe.append(np.array(Test))
        
        #--------------------------------------
    Data.append( (TT, VV, TeTe) )
        
import pickle 
with open('processed_data.pkl', 'wb') as f:
    pickle.dump(Data, f)
```
